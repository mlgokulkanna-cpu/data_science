```python
columns= billings_df.columns
total_count=len(billings_df)
columns_with_min_null=[]
removed=[]
for x in columns:
   
    null_count= (billings_df[x].isnull().sum())
    print(x ,"   ", "Total count:" , total_count,"   ", "NUll :", null_count , end="  |  ")
    percent =(null_count/total_count)*100
    if(percent<25):
        columns_with_min_null.append(x)
    else :
        removed.append(x)
    print()
print(columns_with_min_null)
print(removed)



# print(columns)
```

```python
useable_df=billings_df[columns_with_min_null]
display(useable_df)
```

```python

# Define the columns to clean and handle nulls for categorical data
selected_columns = [
    'Prospect_Outcome', 'Total_Renewal_Score_New', 'Sustainability_Score',
    'Auto_Renewal_Score', 'Status_Scores', 'Anchoring_Score', 'Tenure_Scores',
    'Tenure_Years', 'Current_Auto_Renewal_Flag', 'Proforma_Auto_Renewal',
    'Last_Years_Price', 'Total_Net_Paid', 'Payment_Method', 'Band',
    'Tenure_Group', 'Connection_Group', 'Anchor_Group', 'Payment_Timeframe',
    'Proforma_Audit_Status'
]

# Keep only the selected columns available in the current dataframe
selected_columns = [col for col in selected_columns if col in useable_df.columns]
print('Selected columns available for cleaning:', selected_columns)

# Work on a fresh copy to avoid modifying prior notebook cells indirectly
cleaned_df = useable_df[selected_columns].copy()

# Identify categorical columns and numeric columns among the selected subset
categorical_columns = cleaned_df.select_dtypes(include=['object', 'category','str']).columns.tolist()
numeric_columns = cleaned_df.select_dtypes(include=['int64', 'float64']).columns.tolist()

print('Categorical columns:', categorical_columns)
print('Numeric columns:', numeric_columns)

# Fill missing values for categorical columns with the mode if available, else 'Unknown'
for col in categorical_columns:
    if cleaned_df[col].isnull().any():
        mode_value = cleaned_df[col].mode(dropna=True)
        fill_value = mode_value.iloc[0] if not mode_value.empty else 'Unknown'
        cleaned_df[col].fillna(fill_value, inplace=True)
        

cleaned_df.dropna(subset=['Band','Proforma_Auto_Renewal'], inplace=True)
cleaned_df.fillna({col: 'Unknown' for col in cleaned_df.columns if col in ['Tenure_Group']}, inplace=True)
        

# Preserve the cleaned dataframe for later use
useable_df[selected_columns] = cleaned_df

# Show summary of null counts after cleaning
cleaned_df.isnull().sum()

# cleaned_df.head(20)
```
