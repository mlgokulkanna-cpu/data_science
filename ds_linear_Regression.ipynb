```python
import numpy as np

def train_test_split_manual(X, y, test_size, shuffle=True, random_state=None):

    X = np.array(X)
    y = np.array(y)
          
    n_samples = len(X)
    n_test = int(n_samples * test_size)
    
    indices = np.arange(n_samples)
    
    
    if shuffle:
        if random_state is not None:
            np.random.seed(random_state)
        np.random.shuffle(indices)
    
    test_indices = indices[:n_test]
    train_indices = indices[n_test:]
    
    return X[train_indices], X[test_indices], y[train_indices], y[test_indices]

```

```python
def linear_regression(x,y ,x_test):
    X=np.array(x).flatten()
    X_test=np.array(x_test).flatten()
    y=np.array(y)

    y_mean= np.mean(y)
    X_mean= np.mean(X)
    beta_numerator=0
    beta_denominator=0
    for i in range(X.size):
        beta_numerator += (X[i]-X_mean)*(y[i]-y_mean)
        beta_denominator += (X[i]-X_mean)**2
    beta =(beta_numerator/beta_denominator)
    c= (y_mean) - beta*(X_mean)

    y_pred= beta* X_test + c
    
    return float(beta) , float(c) , y_pred
        
```

```python
X = [[4], [9], [8], [8], [7],[5]]
y = [52, 51, 50, 52, 70 , 60 ]

X_train, X_test, y_train, y_test = train_test_split_manual(X, y, test_size=0.2, random_state=42)

b, c , y_pred = linear_regression(X_train, y_train , X_test)

error = []
for i in range(len(y_test)):
    error.append(y_test[i] - y_pred[i])
print("Test Errors:", error)

print("X_train:", X_train)
print("X_test:", X_test)
print("y_train:", y_train)
print("y_test:", y_test)


y_new = [[6],[6],[5],[9]]
y_new_pred = []
for i in range(len(y_new)):
    
    prediction = (b * y_new[i][0]) + c 
    y_new_pred.append(prediction)

print("Predictions for y_new:", y_new_pred)
```

```python
import numpy as np

class MultiLinearRegression:

    def __init__(self):
        self.beta = None  
        self.const = None 

    def fit(self, X, y):
        
        X = np.array(X)
        y = np.array(y)
        r, c = X.shape
        X_b = np.c_[np.ones((r, 1)), X]
            
        b = np.linalg.pinv((X_b.T).dot( X_b)).dot( (X_b.T .dot(y)))
        self.const = b[0]
        self.beta = b[1:]

    def predict(self, X):
        X = np.array(X)
        return self.const + X @ self.beta      
        
    def get_params(self):
        return {"intercept": self.const, "coefficients": self.beta}
    
```

```python
X=[[4,1,3],[9,2 ,1],[8,3,2],[8,1,1],[7,1,4],[5,0,2]]
y = [52, 51, 50, 52, 70 , 60 ]

X_train, X_test, y_train, y_test = train_test_split_manual(X, y, test_size=0.2, random_state=42)

model = MultiLinearRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_train)
params = model.get_params()

print("Intercept:", params["intercept"])
print("Coefficients:", params["coefficients"])
print("Predictions:", predictions)

```
