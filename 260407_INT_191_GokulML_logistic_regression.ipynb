```python

import numpy as np
import math

class LogisticRegression:

    def __init__(self):
        self.beta = None  
        self.const = None 

    def fit(self, X, y):
        
        X = np.array(X)
        y = np.array(y)
        r, c = X.shape
        X_b = np.c_[np.ones((r, 1)), X]
            
        b = np.linalg.pinv((X_b.T).dot( X_b)).dot( (X_b.T @ y))
                
        self.const = b[0]
        self.beta = b[1:]
        

    def predict(self, X):
        X = np.array(X)
        
        power_expresssion= self.const
        for i in range(len(self.beta)):
            power_expresssion+= X[i]*(self.beta[i])
        return 1/(1+math.exp(-(power_expresssion)))
        
    def get_params(self):
        return {"intercept": self.const, "coefficients": self.beta}
    
```

```python
X = np.array([
    [4, 1, 3],
    [9, 2, 1],
    [8, 3, 2],
    [8, 1, 1],
    [7, 1, 4],
    [5, 0, 2]], dtype='float32')
 
y = np.array([0, 1, 0, 1, 1, 1], dtype='float32')
 
 
split_index = int(len(X) * 0.8)
 
X_train = X[:split_index]
X_test  = X[split_index:]
 
y_train = y[:split_index]
y_test  = y[split_index:]

model= LogisticRegression()
model.fit(X_train,y_train) 
 
result = model.predict([4, 1, 3])
 
if result > 0.5:
  print("category 1" )
  print(result)
else:
  print("category 0")
  print(result)
 
 
```
